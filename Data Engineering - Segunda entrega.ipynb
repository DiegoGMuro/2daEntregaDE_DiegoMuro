```python
import requests
import pandas as pd
```

# URL de la API

```python
url = 'https://tradestie.com/api/v1/apps/reddit'
```

# Encabezados de la solicitud

```python
headers = {"Accept-Encoding": "gzip, deflate"}
```

# Realizar la solicitud GET

```python
response = requests.get(url, headers=headers)
```

# Comprobar el estado de la respuesta

```python
if response.status_code == 200:
    # Si la respuesta es exitosa, obtener los datos en formato JSON
    data = response.json()
    
    # Crear un DataFrame de Pandas para visualizar los datos
    df = pd.DataFrame(data)
    
    # Visualizar el DataFrame
    display(df)
else:
    print("La solicitud no fue exitosa. Código de estado:", response.status_code)
```

# MANIPULACION DE DATOS

```python
# Nuevos datos a insertar (a proposito se duplica ´GAL´ para verificar que no se inserten duplicados)
nuevos_datos = [
    {'no_of_comments': 14, 'sentiment': 'stable', 'sentiment_score': 0.03, 'ticker': 'MELI'},
    {'no_of_comments': 28, 'sentiment': 'stable', 'sentiment_score': 0.09, 'ticker': 'GAL'},
    {'no_of_comments': 8, 'sentiment': 'stable', 'sentiment_score': 0.19, 'ticker': 'GAL'}
]

# Convertir los nuevos datos en un DataFrame
nuevos_registros = pd.DataFrame(nuevos_datos)

# Concatenar y eliminar duplicados basados en 'ticker' (PK)
df = pd.concat([df, nuevos_registros], ignore_index=True).drop_duplicates(subset='ticker')

# Mostrar el DataFrame actualizado
print("DataFrame con los nuevos registros sin duplicados:")
display(df)
```

```python
# Contar la cantidad de Tickers duplicados (Primary Key)
cantidad_duplicados = df.duplicated(subset='ticker').sum()

print("Cantidad de Tickers duplicados:", cantidad_duplicados)
```

# Validador de ´Tickers´ que duplicarian

```python
# Verificar si la nueva fila causaría duplicados en la clave primaria 'ticker'
nuevo_ticker = 'AAPL'  # Valor del nuevo ticker a verificar

if nuevo_ticker in df['ticker'].values:
    print("Advertencia: El nuevo ticker ya existe en el DataFrame. Podría causar una entrada duplicada.")
else:
    print("El nuevo ticker no está duplicado. Puede insertarse con seguridad.")
```

# Tamaño del DATAFRAME

```python
# Mostrar la cantidad de filas y columnas del DataFrame

print("Cantidad de filas:", df.shape[0])
print("Cantidad de columnas:", df.shape[1])
```

# Analisis de informacion del DF

```python
# Suma de 'no_of_comments' por cada 'sentiment'

suma_por_sentiment = df.groupby('sentiment')['no_of_comments'].sum().reset_index()
print("\nSuma de 'no_of_comments' por cada 'sentiment':")
display(pd.DataFrame(suma_por_sentiment))


# Promedio de 'no_of_comments' por cada 'sentiment'

promedio_por_sentiment = df.groupby('sentiment')['no_of_comments'].mean().reset_index()
print("\nPromedio de 'no_of_comments' por cada 'sentiment':")
display(pd.DataFrame(promedio_por_sentiment))
```
